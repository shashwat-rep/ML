Handling missing values

1) using median (for numerical data)

```python
df[numerical_data]= df[numerical_data].fillna(df[numerical_data].median())
```

2. using mode (for categorical data)

```python
for cols in non_numerical_data:
  df[cols] = df[cols].fillna(df[cols].mode()[0])
```

```python
df.isnull().sum()
```

Convert Date (which is a text rn)

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
```

```python
df.drop(columns =['Date'], inplace =True)
```

For categorical rows we use
1) One hot encoding
2) Label encoding

We need to do this because Ml only works on no. so we are gonna need our dataset to contain numerical values only

**One hot encoding** is where there is like no general relationship or sequence in the data. eg- Gender, City, DOB etc ...... if we were to use label encoding here then lelationship would be established like Delhi > Dehradun > Mumbai

```python
df = pd.get_dummies(df, columns=['City'], drop_first=True)
```

**Label encoding** is used where there is relationship between the variables ..... gor eg - Grades etc ...... the realtionship is established as O > A+ > A > B+ > B > C+ > C > F

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['AQI_Bucket'] = le.fit_transform(df['AQI_Bucket'])
```
